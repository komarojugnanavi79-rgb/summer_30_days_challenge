```python
#the index of the largest element.
lst=[1,2,3,4,5]
largest = lst[0]
for num in lst:
  if num >largest:
    largest=num
print("The index of the largest num is:",lst.index(largest))
```

```python
#count how many times the largest element appears
lst=[1,2,3,4,5,5]
largest=lst[0]
count=0
for num in lst:
  if num > largest:
    largest= num
    count = lst.count(largest)
print(lst)
print(largest)
print(count)
```

```python
#the difference between the sum of even numbers and the sum of odd numbers.
lst=[1,2,3,4,5,6,8,9,11]
even_sum=0
odd_sum=0
for num in lst:
  if num % 2==0:
    even_sum+=num
  else:
    odd_sum+=num
diff=abs(even_sum-odd_sum)
print("Sum of even numbers",even_sum)
print("sum of odd numbers",odd_sum)
print(diff)
```

```python
#remove all occurrences of a given
number and print the updated list.
lst=[1,3,2,4,5,4,3,6,7]
key=2
updated_lst=[]
for num in lst:
  if num !=key:
    updated_lst.append(num)
print(updated_lst)
```

```python
#create a new list where each element is doubled and print the result.
lst=[1,2,3,4]
new_lst=[]
for num in lst:
  new_lst.append(num*2)

print(new_lst)
```
