```python
num=int(input("Enter your number:"))

if num % 3 == 0  and num % 5 == 0:
  print("Divisible by both 3 and 5.")
else:
  print("Not divisible.")
```

```python
available_balance=10000
withdrawal_amount=float(input("Enter withdrawal amount:"))
if withdrawal_amount<=available_balance:
  available_balance-=withdrawal_amount
  print("withdrawal successfull! Remaining Balance:",available_balance)
else:
  print("Transaction Rejected! Unsufficient balance.")
```

```python
marks=int(input("Enter Your marks:"))
has_sports_certificate=input("Do you have Sports certificate? (yes/no):").strip().lower() == 'yes'
if marks >= 70 or has_sports_certificate:
  print("Eligible for Admission.")
else:
  print("Not Eligible for Admission.")
```

```python
year=int(input("Enter the year:"))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print(year,"Leap Year.")
else:
  print(year,"Not a leap year.")
```
